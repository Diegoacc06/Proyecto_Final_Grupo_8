```python
import pandas as pd
```

```python
df=pd.read_csv("/workspaces/Proyecto_Final_Grupo_8/pandas/Quiz Cálculo 1- Gobierno 22 de Octubre(1-18).csv")
```

```python
def costo_promedio2(x):
    def crear_funcion_costo(num):
        polinomio= [23,2,num]
        c101=0
        c100=0
        exponente=0
        for k in polinomio:
            c101+=k*(101)**exponente
            c100+=k*(100)**exponente
            exponente+=1
        return (c101-c100)/(101-100)
    return crear_funcion_costo(x[4])
```

```python
df["Valores Reales Punto 2"]=df.apply(costo_promedio2, axis=1 )
```

```python
df
```

```python
def calificacion_P2(x):
    if x[1]==x[5]:
        return 5
    else:
        return 1
```

```python
df["Nota Punto 2"]=df.apply(calificacion_P2, axis=1)
```

```python
df
```

```python
def costo_marginal(x):
    def crear_funcion_costo(num):
        polinomio= [2,2*num]
        c100=0
        exponente=0
        for k in polinomio:
            c100+=k*(100)**exponente
            exponente+=1
        return c100
    return crear_funcion_costo(x[4])
```

```python
df["Valores Reales Punto 3"]= df.apply(costo_marginal, axis=1)
```

```python
df
```

```python
columna_object_idx = 3  # posición de la columna de tipo object

#Se procede a cambiar los valores con comas, por punto

df.iloc[:, columna_object_idx] = (
    df.iloc[:, columna_object_idx]
      .astype(str)
      .str.replace(",", ".", regex=False)
)

#Ahora se cambian todas las columnas a tipo numerico

df.iloc[:, columna_object_idx] = pd.to_numeric(df.iloc[:, columna_object_idx], errors="coerce")
```

```python
def calificacion_P3(x):
    if x[3]==x[9]:
        return 5
    else:
        return 1
```

```python
df["Nota Punto 3"]= df.apply(calificacion_P3, axis=1)
```
