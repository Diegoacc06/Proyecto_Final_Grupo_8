```python
def sumar_digitos(x):
    def sumar(numero):
        num= str(numero)
        suma= 0
        for k in num:
            suma+=int(k)
        return suma
    return sumar(x[0])
```

```python
df["n"]= df.apply(sumar_digitos,axis=1)
```

```python
df.head(5)
```

```python
def costo_promedio(x):
    def crear_funcion_costo(num):
        polinomio= [23,2,num]
        c105=0
        c100=0
        exponente=0
        for k in polinomio:
            c105+=k*(105)**exponente
            c100+=k*(100)**exponente
            exponente+=1
        return (c105-c100)/(105-100)
    return crear_funcion_costo(x[4])
```

```python
df["Valores Reales Punto 1"]=df.apply(costo_promedio, axis=1)
```

```python
df.head(5)
```

```python
def calificacion_P1(x):
    if x[1]==x[5]:
        return 5
    else:
        return 1
```

```python
df["Nota Punto 1"]=df.apply(calificacion_P1, axis=1)
```
