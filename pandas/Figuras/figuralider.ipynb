```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("/workspaces/Proyecto_Final_Grupo_8/pandas/Quiz Cálculo 1- Gobierno 22 de Octubre(1-18).csv")
```

```python
columnas = ['CP1', 'CP2', 'CP3', 'NF']
promedios = [df[col].mean() for col in columnas]

plt.figure(figsize=(8,6))
bars = plt.bar(columnas, promedios, color=['#FF5733', "#F7FF02", '#3357FF', "#33EBFF"], edgecolor='black', linewidth=1.5)
plt.title('Promedios')
plt.ylabel('Promedio')
plt.ylim(0, max(promedios)*1.2)  
plt.show()
```
