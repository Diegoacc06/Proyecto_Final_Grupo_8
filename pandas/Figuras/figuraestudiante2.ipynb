```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("/workspaces/Proyecto_Final_Grupo_8/pandas/Quiz Cálculo 1- Gobierno 22 de Octubre(1-18).csv")
```

```python
p = []
nf = []
for i in range(10, 100, 10):
    p.append(i)
    nf.append(df['NF'].quantile(i/100))

plt.bar(p, nf, color=['red','orange','yellow','green','blue','purple','pink','cyan','magenta'], edgecolor='black')
plt.title('Percentiles')
plt.xlabel('Percentil (%)')
plt.ylabel('NF')
plt.show()
```
