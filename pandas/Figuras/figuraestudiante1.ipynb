```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("/workspaces/Proyecto_Final_Grupo_8/pandas/Quiz Cálculo 1- Gobierno 22 de Octubre(1-18).csv")
```

```python
columnas = ['CP1', 'CP2', 'CP3', 'NF']
Desviaciones = [df[col].std() for col in columnas]

plt.figure(figsize=(8,6))
bars = plt.bar(columnas, Desviaciones, color=['#00FF9F', '#FF00E6', '#0066FF', '#FFD600'], edgecolor='black', linewidth=1.5)
plt.title('Desviaciones')
plt.ylabel('Desviación')
plt.ylim(0, max(Desviaciones)*1.2)  
plt.show()
```
