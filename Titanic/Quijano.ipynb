```python
import pandas as pd
```

```python
df_titanic= pd.read_csv("/workspaces/Proyecto_Final_Grupo_8/Titanic/test.csv")
```

```python
contador_clases= df_titanic["Pclass"].value_counts()
```

```python
print("¿Cuántos pasajeros hay en cada Clase (Pclass)?")
print(f"- Por cada clase hay: {contador_clases}")
```

```python
edad_promedio_sexo = df_titanic.groupby('Sex')['Age'].mean()
```

```python
print("¿Cuál es la edad promedio (Age) de los pasajeros, agrupados por su Sexo?")
print(f"La edad promedio por sexo es: {edad_promedio_sexo}")
```
